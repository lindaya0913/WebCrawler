## 分5個 Dimension

```python
import pandas as pd # 引用套件並縮寫為 pd  
df_split = pd.read_csv('KOKO_split.csv', encoding='utf_8_sig', index_col=0) 
df_split
```

```python
import re
Security_Privacy = [line.strip() for line in open("Security_Privacy.txt", 'r', encoding='utf-8').readlines()]
Practicity = [line.strip() for line in open("Practicity.txt", 'r', encoding='utf-8').readlines()]
Design = [line.strip() for line in open("Design.txt", 'r', encoding='utf-8').readlines()]
Sociality = [line.strip() for line in open("Sociality.txt", 'r', encoding='utf-8').readlines()]
print(Security_Privacy)
print(Practicity)
print(Design)
print(Sociality)
a=[]
for index,review in enumerate(df_split['Section']):
    review = str(review)
    for i in Security_Privacy:
        if review.find(i) != -1:
            review=review.replace(i,"aaa")
    df_split.loc[index,'Section2']=review
    for i in Practicity:
        if review.find(i) != -1:
            review=review.replace(i,"bbb")
    df_split.loc[index,'Section2']=review
    for i in Design:
        if review.find(i) != -1:
            review=review.replace(i,"ccc")
    df_split.loc[index,'Section2']=review
    for i in Sociality:
        if review.find(i) != -1:
            review=review.replace(i,"ddd")
    df_split.loc[index,'Section2']=review
    a.append(re.findall(r"[a-e]{3}",df_split.loc[index,'Section2']))
df_split["Dimension"]=a
```

```python
b=[]
for index_1,review in enumerate(df_split['Section']):
    review = str(review)
    a=[]
    #判斷 review 是否包含 Security_Privacy
    if any(i in review for i in Security_Privacy):
        a.append(1)
    #判斷 review 是否包含 Practicity
    if any(i in review for i in Practicity):
        a.append(2)
    #判斷 review 是否包含 Design
    if any(i in review for i in Design):
        a.append(3)
    #判斷 review 是否包含 Sociality
    if any(i in review for i in Sociality):
        a.append(4)
    b.append(a)
df_split["Dimension2"]=b
```

```python
df_split.to_csv('KOKO_Dimension.csv',encoding='utf_8_sig')
```

## Sentimental analysis Santa_LSTM

```python
import pandas as pd # 引用套件並縮寫為 pd  
df_senti = pd.read_csv('KOKO_Dimension.csv', encoding='utf_8_sig', index_col=0) 
df_senti
```

```python
# !pip install opencc
from opencc import OpenCC
review_char = []
cc = OpenCC('tw2sp')
for index,review in enumerate(df_senti['Section']):
    review = str(review)
    review_char.append(cc.convert(review))
#     print(cc.convert(review))
```

```python
# !pip3 install --upgrade paddle
import paddlehub as hub
# load model
senta = hub.Module(name="senta_lstm")
# load skep model
# senta = hub.Module(name="ernie_skep_sentiment_analysis")

# text=[]
sentiment_key=[]
positive_probs=[]
negative_probs=[]

# text
for index,review in enumerate(review_char):
    review = str(review)
    test_text = [review]
    # classify
    results = senta.sentiment_classify(data={"text": test_text})
    # result
    for result in results:
#         text.append(result['text'])
        sentiment_key.append(result['sentiment_key'])
        positive_probs.append(result['positive_probs'])
        negative_probs.append(result['negative_probs'])

df_senti["Sentiment_key"]=sentiment_key
df_senti["Positive_probs"]=positive_probs
df_senti["Negative_probs"]=negative_probs
```

```python
df_senti.to_csv('KOKO_sentimental.csv',encoding='utf_8_sig')
```

## Sentimental strength level

```python
import pandas as pd # 引用套件並縮寫為 pd  
df_level = pd.read_csv('KOKO_sentimental.csv', encoding='utf_8_sig', index_col=0) 
df_level
```

```python
level_0 = [line.strip() for line in open("level_0.txt", 'r', encoding='utf-8').readlines()]
level_1 = [line.strip() for line in open("level_1.txt", 'r', encoding='utf-8').readlines()]
level_2 = [line.strip() for line in open("level_2.txt", 'r', encoding='utf-8').readlines()]
level_3 = [line.strip() for line in open("level_3.txt", 'r', encoding='utf-8').readlines()]
print(level_0)
print(level_1)
print(level_2)
print(level_3)
b=[]
for index,review in enumerate(df_level['Section']):
    review = str(review)
    a=[]
    for i in level_0:
        if review.find(i) != -1:
            a.append(2)
    for i in level_1:
        if review.find(i) != -1:
            a.append(3)
    for i in level_2:
        if review.find(i) != -1:
            a.append(4)
    for i in level_3:
        if review.find(i) != -1:
            a.append(5)
    if review.find("最近")!= -1:
        a.append(-4)
    if review.find("最後")!= -1:
        a.append(-4)
    if review.find("滿意")!= -1:
        a.append(-3)
    if review.find("不滿")!= -1:
        a.append(-3)
    if a == []:
        a=[1]
    b.append(a)
df_level["Level"]=b
df_level
```

```python
df_level.to_csv('KOKO_level.csv',encoding='utf_8_sig')
```
